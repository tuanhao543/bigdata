```python
import pandas as pd

# Tính toán số lượng cột cần tạo
max_genres_count = df_cleaned['genres'].str.count('|').max() + 1

# Tạo danh sách cột mới
genres_columns = [f'genres{i+1}' for i in range(max_genres_count)]

# Tạo ra đúng số lượng cột cần thiết với giá trị None
for col in genres_columns:
    df_cleaned[col] = None

# Thêm giá trị cho từng cột
for i in range(max_genres_count):
    # Kiểm tra nếu chỉ mục vượt quá số lượng cột thực tế
    if i < len(df_cleaned['genres'].str.split('|', expand=True).columns):
        df_cleaned[genres_columns[i]] = df_cleaned['genres'].str.split('|', expand=True).iloc[:, i]

# Xóa những cột có toàn bộ giá trị là None
df_cleaned = df_cleaned.dropna(axis=1, how='all')

```

```python
df_cleaned[['movieId', 'title', 'year', 'genres1', 'genres2', 'genres3', 'genres4', 'genres5', 'genres6', 'genres7', 'genres8', 'genres9', 'genres10']]
```

```python
#  sử dụng thư viện scikit-learn để vector hóa TF-IDF trên các xâu ký tự trong cột "clean_title"
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(ngram_range=(1,2))

tfidf = vectorizer.fit_transform(df_cleaned["genres1"])
```

```python
tfidf
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def search(genres):
    # chuyen thanh vecto TF-IDF
    query_vec = vectorizer.transform([genres])
    # tinh do tuong dong giua title can tim voi title khac
    similarity = cosine_similarity(query_vec, tfidf).flatten()
    # xac dinh 5 phim tuong dong nhat
    indices = np.argpartition(similarity, -5)[-5:]
    results = df_cleaned.iloc[indices].iloc[::-1]
    
    return results
```

```python
import ipywidgets as widgets
from IPython.display import display
# tao o tim kiem
movie_input = widgets.Text(
    value='',
    description='Movie Genres:',
    disabled=False
)
# hien thi ket qua 
movie_list = widgets.Output()
# ham xu ly khi thay doi nhap lieu
def on_type(data):
    with movie_list:
        movie_list.clear_output()
        genres = data["new"]
        if len(genres) > 5:
            display(search(genres))

movie_input.observe(on_type, names='value')


display(movie_input, movie_list)
```
